```python
def printbricks(layer = 1, height = 1):
    bricksleft = printbricksleft (layer, height)
    bricksright = printbricksright(layer)
    print(bricksleft, "  ", bricksright, sep="")

def printbricksright(layer = 1):
    bricksright = "#" * layer
    return bricksright  

def printbricksleft(layer = 1, height = 1):
    spaceleft = " " * (height  -layer)
    brickleft = "#" * layer
    spacedbricksleft = spaceleft + brickleft
    return spacedbricksleft

height = int(input("please define the number of layers for your pyramid: "))
layer = 1
while layer <= height:
    printbricks(layer, height)
    layer = layer +1
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
def binary_search_recursive(arr, x, low, high):
    if low <= high:
        mid = (low + high) // 2

        # Eğer x, ortanca elemana eşitse, bulunduğumuz indeksi döndür
        if arr[mid] == x:
            return mid

        # Eğer x, ortanca elemandan küçükse, sağ tarafı kontrol et
        elif arr[mid] < x:
            return binary_search_recursive(arr, x, mid + 1, high)

        # Eğer x, ortanca elemandan büyükse, sol tarafı kontrol et
        else:
            return binary_search_recursive(arr, x, low, mid - 1)

    # Eleman bulunamadıysa -1 döndür
    return -1

# Örnek kullanım
arr = list(range(60000000)) 
x = 399

# Fonksiyonu çağırarak arama yapın
result = binary_search_recursive(arr, x, 0, len(arr) - 1)

if result != -1:
    print(f"Eleman {x} listede bulundu, indeks: {result}")
else:
    print(f"Eleman {x} listede bulunamadı.")
```
